```python
#34 Classify Movies Based on Ratings [Excellent, Good and Average]
def fn(rating):
    if rating >=9:
        return "Excellent"
    if rating >=7:
        return "Good"
    else:
        return "Average"
    
```

```python
#37 Find Unique Values From Genre
list=[]
for i in df['Genre']:
    list.append(i.split(','))
```

```python
newlist=[]
for j in list:
    for k in j:
        newlist.append(k)
```

```python
uni_list= []
for x in newlist:
    if x not in uni_list:
        uni_list.append(x)
        
```
